```python
import numpy as np
import pandas as pd


def part1(fileData):  # Most expensive car
    print("******************\nPart 1")
    print(fileData[fileData['price'] == max(
        fileData['price'])].to_string(index=False).strip())


def part2(fileData):  # Calculate total cars per manufacturer
    print("******************\nPart 2")
    for i in fileData['company'].unique():
        print(i, end=" ")
        print(len(fileData[fileData['company'] == i]))


def part3(fileData):  # details of vehicles against Toyota manufacturer
    print("******************\nPart 3")
    print(fileData.groupby('company').get_group('toyota'))


def part4(fileData):  # Arrange the cars according to the prices and print first 5 set
    print("******************\nPart 4")
    print(fileData.sort_values(by=['price'], ascending=False).head(5))


fileData = pd.read_csv("automobile_data.csv")

part1(fileData)
part2(fileData)
part3(fileData)
part4(fileData)
```
