```python
import numpy as np
import pandas as pd


def part1(fileData):  # print First 5 Set
    print("************************\nPart1 :")
    print(fileData.head(6))


def part2(fileData):  # print Last 5 Set
    print("************************\nPart2 :")
    print(fileData.tail(5))


def part3(fileData):  # print Columns Label
    print("************************\nPart3 :")
    print(fileData.columns)


def part4(fileData):  # print Unique Value for each columns
    print("************************\nPart4 :")
    column_name = ["index", "company", "body-style", "wheel-base", "length", "engine-type",
                   "num-of-cylinders", "horsepower", "average-mileage", "price"]

    for data in fileData:
        print("------------------")
        print(data)
        print(fileData[data].nunique())


def part5(fileData):  # print Count Of Null for each columns
    print("************************\nPart5 :")
    column_name = ["index", "company", "body-style", "wheel-base", "length", "engine-type",
                   "num-of-cylinders", "horsepower", "average-mileage", "price"]

    for data in fileData:
        print("------------------")
        print(data)
        print(fileData[data].isnull().sum())


def part6(fileData):  # remove Unnecessary Col
    print("************************\nPart 6:")
    new_data = fileData.drop(['company'], axis=1)
    print(new_data)


def part7(fileData):  # Drop Null Value
    print("************************\nPart 7 : ")
    fileData.dropna(subset=fileData.columns, inplace=True)
    print(fileData)


fileData = pd.read_excel("automobile_data.xlsx")
part1(fileData)
part2(fileData)
part3(fileData)
part4(fileData)
part5(fileData)
part6(fileData)
part7(fileData)
```
