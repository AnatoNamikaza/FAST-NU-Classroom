```python
import matplotlib.pyplot as plt # use to draw graph
import numpy as np # use to store computed data
import pandas as pd # use to read file


Record = pd.read_csv('fifa_data.csv')

plt.boxplot(np.array([Record[(Record['Weight']>='125lbs') & (Record['Weight']<'150lbs')]['Weight'].count(), Record[(Record['Weight']>='150lbs') & (Record['Weight']<'175lbs')]['Weight'].count(), Record[(Record['Weight']>='175lbs') & (Record['Weight']<'200lbs')]['Weight'].count(), Record[(Record['Weight']<'125lbs') ]['Weight'].count(), Record[(Record['Weight']>'200lbs') ]['Weight'].count()]))
plt.show()

plt.pie(np.array([Record[(Record['Weight']>='125lbs') & (Record['Weight']<'150lbs')]['Weight'].count() / Record['Weight'].count() * 100, Record[(Record['Weight']>='150lbs') & (Record['Weight']<'175lbs')]['Weight'].count()/ Record['Weight'].count() * 100, Record[(Record['Weight']>='175lbs') & (Record['Weight']<'200lbs')]['Weight'].count() / Record['Weight'].count() * 100, Record[(Record['Weight']<'125lbs') ]['Weight'].count() / Record['Weight'].count() * 100, Record[(Record['Weight']>'200lbs') ]['Weight'].count() / Record['Weight'].count() * 100]),labels = np.array(['125-150','150-75','175-200','> 200','< 125']));
plt.show()

plt.bar(np.array(['125-150','150-75','175-200','> 200','< 125']),np.array([Record[(Record['Weight']>='125lbs') & (Record['Weight']<'150lbs')]['Weight'].count(), Record[(Record['Weight']>='150lbs') & (Record['Weight']<'175lbs')]['Weight'].count(), Record[(Record['Weight']>='175lbs') & (Record['Weight']<'200lbs')]['Weight'].count(), Record[(Record['Weight']<'125lbs') ]['Weight'].count(), Record[(Record['Weight']>'200lbs') ]['Weight'].count()]), width=0.9, align='center')
plt.show()
```
